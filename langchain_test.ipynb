```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint

# Hugging Face Inference API (NO local model download)
llm = HuggingFaceEndpoint(
    endpoint_url="Qwen/Qwen3-Coder-30B-A3B-Instruct",
    task="text-generation",
    max_new_tokens=1024,
    do_sample=False,
)

chat_model = ChatHuggingFace(llm=llm)

# ---- Chat-only pseudocode pipeline ----
def generate_pseudocode(control_narrative: str):

    messages = [
        {
            "role": "system",
            "content": (
                "You are an expert software engineer. "
                "Convert the given control narrative into clear, structured pseudocode. "
                "Use IF, ELSE, WHILE, FOR, FUNCTION, and comments."
                "And no other text beyond the pseudocode."
            )
        },
        {
            "role": "user",
            "content": control_narrative
        }
    ]

    response = chat_model.invoke(messages)
    return response.content


# ---- Example Input ----
control_text = """
When the start button is pressed, the motor should turn ON.
If temperature exceeds 80 degrees, turn OFF the motor.
Otherwise keep monitoring the temperature every 5 seconds.
"""

# ---- Run ----
print(generate_pseudocode(control_text))
```
